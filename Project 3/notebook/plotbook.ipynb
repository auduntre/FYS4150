```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
ls ../positions/
```

```python
plt.figure(figsize=(12, 12))
plt.title("Velocity Verlet method over 10 years")
plt.plot([0.0], [0.0] ,"o" ,label="Initial Sun")

for i in range(2, 4):
    x = np.loadtxt("../positions/positions" + str(10**i) + ".txt")
    
    # make sure there is no z component over the plane
    assert x[2, :].all() == 0
    
    
    plt.plot(x[0, :], x[1, :], label = "Timesteps: " + str(10**i))

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
plt.figure(figsize=(12, 12))
plt.title("Euler's method over 10 years")
plt.plot([0.0], [0.0] ,"o" ,label="Initial Sun")

plt.xlim([-3.0, 3.0])
plt.ylim([-3.0, 3.0])

for i in range(2, 6):
    x = np.loadtxt("../positions/positions" + str(10**i) + ".txt")
    
    # make sure there is no z component over the plane
    assert x[2, :].all() == 0
    
    
    plt.plot(x[0, :], x[1, :], label = "Timesteps: " + str(10**i))

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
plt.figure(figsize=(12, 12))
plt.title("Beta plot over 5 years")
plt.plot([0.0], [0.0] ,"o" ,label="Initial Sun")

plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])

betas = ["2.00", "2.50", "2.75", "2.90", "2.95", "3.00"]

for beta in betas:
    x = np.loadtxt("../positions/beta" + beta + "0000.txt")
    
    # make sure there is no z component over the plane
    assert x[2, :].all() == 0
    
    plt.plot(x[0, :], x[1, :], label = "Beta: " + beta)

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
